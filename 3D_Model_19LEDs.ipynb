19LEDs unit is centered at (0,0) with 735 in the center.
Under section 'Calculating irradianice' it is posible to change a distance z_distance, half angle psi, Iro intensity of LEDs.
Irradiance is calculated in 200*200 plane, but is option to change a size of target plane --> to do that change number in  x_numpy variable and  y_numpy variable.
Output is an interactiv 3D model in x,y,z coordinates with option:rotating, zooming.
Positions of LEDs are plotted in x,y coordinates, find below 3D model.

```
##### 19LEDs unit, 3D model ### 
import numpy as np
import matplotlib.pyplot as plt

#LEDs array coordinates 
xlist_1 = [15.89, 15.89, 15.89, 23.27, 23.27, 27.89, 29.89, 29.89, 36.89, 36.89,
           36.89, 43.89, 43.89, 45.86, 51.41, 51.41, 57.89, 57.89, 57.89]
ylist_1 = [8.405, 26.405, 44.405, 17.659, 35.405, 26.405, 8.405, 44.405, 17.695,
           26.405, 35.405, 8.405, 44.405, 26.405, 35.405, 17.695, 8.405, 26.405, 44.405]

# LEDs coordinates centered at (0,0)
xlist0 = []
ylist0 = []
x1 = 36.89 # moving LEDs unit - x axis
y1 = 26.405 # moving LEDs unit - y axis
for i in range(len(xlist_1)):
  xlist0.append(xlist_1[i]-x1)
  ylist0.append(ylist_1[i]-y1)

# Gettting a points in a target plane (xp,yp) in mm
point = 30
x_numpy = np.linspace(-100,100,point) #dimension of target plane
y_numpy = np.linspace(-100,100,point) #dimension of target plane
xplist = []
yplist = []

for x in x_numpy:
  for y in y_numpy:
    xplist.append(x)
    yplist.append(y)

#####Calculating irradiance#######
intensity_point = []
z_distance = 150
#psi angle ( psi is angle at half of max intensity)
psi_angle = 90
psi = psi_angle * np.pi/180 # convert into radians
m = -( np.log(2) / np.log(np.cos(psi))) 
Ir0 = 20 #(mW)   # intensity


inten_sum_point = [] #empty list for saving results
for xp,yp in zip(xplist,yplist):
  intensity_point = []
  for x,y in zip(xlist0,ylist0):    
    E_point = ((z_distance**(m+1))*Ir0) / (((xp-x)**2) + ((yp-y)**2) + (z_distance**2))**((m+3)/2)
    intensity_point.append(E_point) 
  inten_sum_point.append(sum(intensity_point))

import plotly.graph_objects as go
import plotly.offline as go_offline
#prepare data in 
nten_numpy = np.array(inten_sum_point)
Z =  nten_numpy.reshape(point,point)
fig=go.Figure(data = go.Surface(z=Z, x = x_numpy, y = y_numpy,opacity=0.9, colorscale='Viridis'))
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                             highlightcolor="limegreen", project_z=True))
fig.update_layout(title=(f" 19LEDs unit, distance: {z_distance}(mm) and half angle:{psi_angle}deg"))  
fig.show()

fig = go.Figure( data = go.Scatter(x = xlist0, y =ylist0, opacity=0.9, mode='markers'))
fig.update_layout(title = "LEDs positions")
fig.show()


```

The below is code for 3D model where 6 units are placed in regular hexagon. 
A regular hexagon is with six equal sides, the side is named as "a" in the code.Idea is to change a lenght of side "a" in term of better uniformity.
Output is an 3D model of irradiance and calculations such as Emin, E avarage, factor uniformity "u".

```
#################### The six 19LEDs unit placed in regular hexagon shape ########################
import numpy as np
import matplotlib.pyplot as plt
import statistics
# distance and angle
z_distance = 150
psi_angle = 90 #psi angle ( psi is angle at half of max intensity)
psi = psi_angle * np.pi/180 # convert into radians

#LEDs array coordinates
xlist_1 = [15.89, 15.89, 15.89, 23.27, 23.27, 27.89, 29.89, 29.89, 36.89, 36.89,
           36.89, 43.89, 43.89, 45.86, 51.41, 51.41, 57.89, 57.89, 57.89]
ylist_1 = [8.405, 26.405, 44.405, 17.659, 35.405, 26.405, 8.405, 44.405, 17.695,
           26.405, 35.405, 8.405, 44.405, 26.405, 35.405, 17.695, 8.405, 26.405, 44.405]

#xlist_1 = [15.89, 15.89, 57.89, 57.89]
#ylist_1 = [8.405, 44.405, 8.405, 44.405]

# LEDs coordinates centered at (0,0)
#empty lists for saving LEDs coordinates
xlist_11 = [] # x coordinates for 1 unit
xlist_22 = [] # x coordinates for 2 unit
xlist_33 = [] # x coordinates for 3 unit
xlist_44 = [] # x coordinates for 4 unit
xlist_55 = [] # x coordinates for 5 unit
xlist_66 = [] # x coordinates for 6 unit

ylist_11 = [] # y coordinates for 1 unit
ylist_22 = [] # y coordinates for 2 unit
ylist_33 = [] # y coordinates for 3 unit
ylist_44 = [] # y coordinates for 4 unit
ylist_55 = [] # y coordinates for 5 unit
ylist_66 = [] # y coordinates for 6 unit

# LEDs coordinates centered at (0,0)
xlist_0 = []
ylist_0 = []
x1 = 36.89 # moving LEDs unit - x axis
y1 = 26.405 # moving LEDs unit - y axis
for i in range(len(xlist_1)):
  xlist_0.append(xlist_1[i]-x1)
  ylist_0.append(ylist_1[i]-y1)

####getting coordinates for a 6 units in regular hexagon shape
a = 116 # distance from (0,0)(mm)
r = (np.sqrt(3)*a)/2
for i in range(len(xlist_0)):
  xlist_11.append(xlist_0[i]+a)
  xlist_22.append(xlist_0[i]+(a/2))
  xlist_33.append(xlist_0[i]-(a/2))
  xlist_44.append(xlist_0[i]-a)
  xlist_55.append(xlist_0[i]-(a/2))
  xlist_66.append(xlist_0[i]+(a/2))

  ylist_11.append(ylist_0[i])
  ylist_22.append(ylist_0[i]-r)
  ylist_33.append(ylist_0[i]-r)
  ylist_44.append(ylist_0[i])
  ylist_55.append(ylist_0[i]+r)
  ylist_66.append(ylist_0[i]+r)
  
xlist_11.extend(xlist_22)
xlist_11.extend(xlist_33)
xlist_11.extend(xlist_44)
xlist_11.extend(xlist_55)
xlist_11.extend(xlist_66)

ylist_11.extend(ylist_22)
ylist_11.extend(ylist_33)
ylist_11.extend(ylist_44)
ylist_11.extend(ylist_55)
ylist_11.extend(ylist_66)


# Gettting a points in a target plane (xp,yp) in mm
point= 60 # number of points b*b
x_numpy = np.linspace(-75,75,point) # target plane -  x axis dimension
y_numpy = np.linspace(-75,75,point) # target plane - y axis dimension
xplist = []
yplist = []

for x in x_numpy:
  for y in y_numpy:
    xplist.append(x)
    yplist.append(y)

#####Calculating irradiance#######
intensity_point = []
m = -( np.log(2) / np.log(np.cos(psi))) 
Ir0 = 200 #(mW)   # intensity


inten_sum_point = [] #empty list for saving results
for xp,yp in zip(xplist,yplist):
  intensity_point = []
  for x,y in zip(xlist_11,ylist_11):    
    E_point = ((z_distance**(m+1))*Ir0) / (((xp-x)**2) + ((yp-y)**2) + (z_distance**2))**((m+3)/2)
    intensity_point.append(E_point) 
  inten_sum_point.append(sum(intensity_point))


#statistics
import statistics
minim = min(inten_sum_point)
maxx = max(inten_sum_point)
print(f"1.Max E : {maxx} (Watt per m2)")
print (f"2.Min E: {minim} (Watt per m2) ")
omjer = minim / maxx
print("3.E min/E max:", omjer)
E_avarage = (sum(inten_sum_point)) / (len(inten_sum_point))
print ("4.E_avarage:" , E_avarage)
u = minim / E_avarage            
print("5.Uniformity of E , u = Emin/Evarage, (the best is when u>0.8):", u)


import plotly.graph_objects as go
import plotly.offline as go_offline
#prepare data in 
nten_numpy = np.array(inten_sum_point)
Z =  nten_numpy.reshape(point,point) #reshaping list to array shape(b , b)
fig=go.Figure(data = go.Surface(z=Z, x = x_numpy, y = y_numpy,opacity=0.9, colorscale='Viridis'))
                      
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                             highlightcolor="limegreen", project_z=True))
fig.update_layout(title=(f" 19 LEDs unit , distance: {z_distance}(mm) and half angle:{psi_angle}deg"))  
fig.show()

fig = go.Figure( data = go.Scatter(x = xlist_11, y =ylist_11, opacity=0.9, mode='markers'))
fig.show()
```


Below is code for calculating the best position for four 19LEDs unit. 

At beginning, the 19LEDs are arranged around (0,0). Code calculates a irradiance over the target plane, if u= minimum irradiance/ avarage irradiance is less than 0.8, each UV lamp will be moved 1mm in x direction and 1mm in y direction. Finally, when u >= 0.8, calculating will be stopped and 3D model will be given as an output. Positions of UV units are given in x,y plot.

```
# 4  *  19 Units , distance between boards
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt


# Unit 1, coordinates 
xlist_1 = [15.89, 15.89, 15.89, 23.27, 23.27, 27.89, 29.89, 29.89, 36.89, 36.89,
           36.89, 43.89, 43.89, 45.86, 51.41, 51.41, 57.89, 57.89, 57.89]
ylist_1 = [8.405, 26.405, 44.405, 17.659, 35.405, 26.405, 8.405, 44.405, 17.695,
           26.405, 35.405, 8.405, 44.405, 26.405, 35.405, 17.695, 8.405, 26.405, 44.405]

# distance and angle
point = 30 # number of points b*b
x_numpy = np.linspace(-100,100,point)
y_numpy = np.linspace(-100,100,point)
xplist = []
yplist = []

for x in x_numpy:
  for y in y_numpy:
    xplist.append(x)
    yplist.append(y)

#####Calculating irradiance#######

z_distance = 150
psi_angle = 90 #psi angle ( psi is angle at half of max intensity)
psi = psi_angle * np.pi/180 # convert into radians
m = -( np.log(2) / np.log(np.cos(psi))) 
Ir0 = 20 #(mW)   # intensity

u=0
a = 0 # distance between units(mm) x axis
c = 0 # distance between units(mm) y axis
while u<0.9:
  #getting coordinates for a four units
  
  xlist_11 = [] # x coordinates for 1 unit
  xlist_22 = [] # x coordinates for 2 unit
  xlist_33 = [] # x coordinates for 3 unit
  xlist_44 = [] # x coordinates for 4 unit

  ylist_11 = [] # y coordinates for 1 unit
  ylist_22 = [] # y coordinates for 2 unit
  ylist_33 = [] # y coordinates for 3 unit
  ylist_44 = [] # y coordinates for 4 unit

  for i in range(len(xlist_1)):
    xlist_11.append(xlist_1[i]+a)
    xlist_22.append(xlist_1[i]+a)
    xlist_33.append((xlist_1[i]-74)-a)
    xlist_44.append((xlist_1[i]-74)-a)

    ylist_11.append(ylist_1[i]+a)
    ylist_22.append((ylist_1[i]-74)-a)
    ylist_33.append((ylist_1[i]-74)-a)
    ylist_44.append(ylist_1[i]+a)
  
  xlist_11.extend(xlist_22)
  xlist_11.extend(xlist_33)
  xlist_11.extend(xlist_44)
  ylist_11.extend(ylist_22)
  ylist_11.extend(ylist_33)
  ylist_11.extend(ylist_44)

  inten_sum_point = [] #empty list for saving results
  for xp,yp in zip(xplist,yplist):
    intensity_point = []
    for x,y in zip(xlist_11,ylist_11):    
      E_point = ((z_distance**(m+1))*Ir0) / (((xp-x)**2) + ((yp-y)**2) + (z_distance**2))**((m+3)/2)
      intensity_point.append(E_point) 
    inten_sum_point.append(sum(intensity_point))
  minim = min(inten_sum_point)

  E_avarage = (sum(inten_sum_point)) / (len(inten_sum_point))

  u = minim / E_avarage            
  print("5.Uniformity of E , u = Emin/Evarage, (the best is when u>0.8):", u)

  a+=1
  print(f"Distance between the units:{a*2} (mm)" )
  
import plotly.graph_objects as go
import plotly.offline as go_offline
#prepare data in 
nten_numpy = np.array(inten_sum_point)
Z =  nten_numpy.reshape(point,point) #reshaping list to array shape(b , b)
fig=go.Figure(data = go.Surface(z=Z, x = x_numpy, y = y_numpy,opacity=0.9, colorscale='Viridis'))
                      
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                             highlightcolor="limegreen", project_z=True))
#fig.update_traces(data = go.Scatter(x = xlist_11, y =ylist_11, mode='markers'))


fig.update_layout(title=(f" 19 LEDs unit , distance: {z_distance}(mm) and half angle:{psi_angle}deg"))  
fig.show()


fig = go.Figure( data = go.Scatter(x = xlist_11, y =ylist_11, opacity=0.9, mode='markers'))
fig.show() 
```
