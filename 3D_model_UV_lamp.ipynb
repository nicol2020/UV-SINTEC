# **3D model of UV lamp **

The (x, y)coordinates of each diode are taken from pcb board and saved to  xlist and ylist lists. The LEDs are saved in order:

310_2 = (xlist[0], ylist[0])

340 = (xlist[1], ylist[1])

310_3 = (xlist[2], ylist[2])

370 = (xlist[3], ylist[3])

325 = (xlist[4], ylist[4])

405 = (xlist[5], ylist[5])

310_5 = (xlist[6], ylist[6])

395 = (xlist[7], ylist[7])

385 = (xlist[8], ylist[8])

365 = (xlist[9], ylist[9])

310_4 = (xlist[10], ylist[10])

280 =  (xlist[11], ylist[11])

UV lamp is centered at (0,0) with 310_5 in the center.
Under section 'Calculating irradianice' it is posible to change a distance z_distance, half angle psi, Iro intensity of LEDs.
Irradiance is calculated in 200*200 plane. 
Output is an interactiv 3D model in x,y,z coordinates with option:rotating, zooming.
Positions of LEDs are plotted in x,y coordinates, find below 3D model.


```
####################  UV lamp   ########################
import numpy as np
import matplotlib.pyplot as plt


#LEDs array coordinates
xlist = [9.5, 18.05, 26.41, 13.83, 21.60, 9.9, 18, 25.09, 13.39, 21.9, 9.59, 17.88, 26.18]
ylist = [9.5, 8.15, 9.5, 13.94, 14.44, 17.82, 18, 17.37, 22.26, 21.84, 25.84, 26.19, 25.90]

# LEDs coordinates centered at (0,0)
xlist0 = []
ylist0 = []
b = 18 
for i in range(len(xlist)):
  xlist0.append(xlist[i]-b)
  ylist0.append(ylist[i]-b)

# Gettting a points in a target plane (xp,yp) in mm
a = 30
x_numpy = np.linspace(-100,100,a) #dimension of target plane
y_numpy = np.linspace(-100,100,a) #dimension of target plane
xplist = []
yplist = []

for x in x_numpy:
  for y in y_numpy:
    xplist.append(x)
    yplist.append(y)

#####Calculating irradiance#######
intensity_point = []
z_distance = 150
#psi angle ( psi is angle at half of max intensity)
psi_angle = 90
psi = psi_angle * np.pi/180 # convert into radians
m = -( np.log(2) / np.log(np.cos(psi))) 
Ir0 = 20 #(mW)   # intensity


inten_sum_point = [] #empty list for saving results
for xp,yp in zip(xplist,yplist):
  intensity_point = []
  for x,y in zip(xlist0,ylist0):    
    E_point = ((z_distance**(m+1))*Ir0) / (((xp-x)**2) + ((yp-y)**2) + (z_distance**2))**((m+3)/2)
    intensity_point.append(E_point) 
  inten_sum_point.append(sum(intensity_point))

import plotly.graph_objects as go
import plotly.offline as go_offline
#prepare data in 
nten_numpy = np.array(inten_sum_point)
Z =  nten_numpy.reshape(a,a)
fig=go.Figure(data = go.Surface(z=Z, x = x_numpy, y = y_numpy,opacity=0.9, colorscale='Viridis'))
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                             highlightcolor="limegreen", project_z=True))
fig.update_layout(title=(f" UV LEDs unit, distance: {z_distance}(mm) and half angle:{psi_angle}deg"))  
fig.show()

fig = go.Figure( data = go.Scatter(x = xlist0, y =ylist0, opacity=0.9, mode='markers'))
fig.update_layout(title = "LEDs positions for UV lamp")
fig.show()
```

Below is code for calculating the best position for four Uv units. 

At beginning, the UV lamps are arranged around (0,0). Code calculates a irradiance over the target plane, if u= minimum irradiance/ avarage irradiance is less than 0.8, each UV lamp will be moved 1mm in x direction and 1mm in y direction. Finally, when u >= 0.8, calculating will be stopped and 3D model will be given as an output. Positions of UV units are given in x,y plot.

```
import numpy as np
import matplotlib.pyplot as plt

#LEDs array coordinates
xlist_1 = [9.5, 18.05, 26.41, 13.83, 21.60, 9.9, 18, 25.09, 13.39, 21.9, 9.59, 17.88, 26.18]
ylist_1 = [9.5, 8.15, 9.5, 13.94, 14.44, 17.82, 18, 17.37, 22.26, 21.84, 25.84, 26.19, 25.90]


# distance and angle
b = 30 # number of points b*b
x_numpy = np.linspace(-100,100,b)
y_numpy = np.linspace(-100,100,b)
xplist = []
yplist = []

for x in x_numpy:
  for y in y_numpy:
    xplist.append(x)
    yplist.append(y)

#####Calculating irradiance#######
#intensity_point = []
z_distance = 150
psi_angle = 90 #psi angle ( psi is angle at half of max intensity)
psi = psi_angle * np.pi/180 # convert into radians
m = -( np.log(2) / np.log(np.cos(psi))) 
Ir0 = 1 #(mW)   # intensity

u=0
a = 0 # distance between units(mm) x axis
c = 0 # distance between units(mm) y axis
while u<0.8:
  #getting coordinates : 4 units
  
  xlist_11 = [] # x coordinates for 1 unit
  xlist_22 = [] # x coordinates for 2 unit
  xlist_33 = [] # x coordinates for 3 unit
  xlist_44 = [] # x coordinates for 4 unit

  ylist_11 = [] # y coordinates for 1 unit
  ylist_22 = [] # y coordinates for 2 unit
  ylist_33 = [] # y coordinates for 3 unit
  ylist_44 = [] # y coordinates for 4 unit

  for i in range(len(xlist_1)):
    xlist_11.append(xlist_1[i]+a)
    xlist_22.append(xlist_1[i]+a)
    xlist_33.append((xlist_1[i]-36)-a)
    xlist_44.append((xlist_1[i]-36)-a)

    ylist_11.append(ylist_1[i]+a)
    ylist_22.append((ylist_1[i]-36)-a)
    ylist_33.append((ylist_1[i]-36)-a)
    ylist_44.append(ylist_1[i]+a)
  
  xlist_11.extend(xlist_22)
  xlist_11.extend(xlist_33)
  xlist_11.extend(xlist_44)
  ylist_11.extend(ylist_22)
  ylist_11.extend(ylist_33)
  ylist_11.extend(ylist_44)

  inten_sum_point = [] #empty list for saving results
  for xp,yp in zip(xplist,yplist):
    intensity_point = []
    for x,y in zip(xlist_11,ylist_11):    
      E_point = ((z_distance**(m+1))*Ir0) / (((xp-x)**2) + ((yp-y)**2) + (z_distance**2))**((m+3)/2)
      intensity_point.append(E_point) 
    inten_sum_point.append(sum(intensity_point))
  minim = min(inten_sum_point)

  E_avarage = (sum(inten_sum_point)) / (len(inten_sum_point))

  u = minim / E_avarage            
  print("5.Uniformity of E , u = Emin/Evarage, (the best is  u>0.8):", u)

  a+=1
  print(f"Distance between UV units :{a*2} (mm)")

#plotting results  
import plotly.graph_objects as go
import plotly.offline as go_offline
nten_numpy = np.array(inten_sum_point)
Z =  nten_numpy.reshape(b,b) #reshaping list to array shape(b , b)
fig=go.Figure(data = go.Surface(z=Z, x = x_numpy, y = y_numpy,opacity=0.9, colorscale='Viridis'))                     
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                             highlightcolor="limegreen", project_z=True))
fig.update_layout(title=(f" UV unit , distance: {z_distance}(mm) and half angle:{psi_angle}deg"))  
fig.show()


fig = go.Figure( data = go.Scatter(x = xlist_11, y =ylist_11, opacity=0.9, mode='markers'))
fig.update_layout(title = "LEDs positions for UV lamp")
fig.show()
```
